```python
import sqlite3
import pandas as pd
```

```python
# Connect to SQLite database
conn = sqlite3.connect('database.sqlite')
cursor=conn.cursor()
```

```python
# Extract data and load into DataFrames
business_rules_df = pd.read_sql_query("SELECT * FROM BusinessRules;", conn)
business_rules_df.head()
```

```python
business_rules_df.tail()
```

```python
business_rules_df.info()
```

```python
# Data Transformation Example: Convert 'ImportDate' to datetime
business_rules_df['ImportDate'] = pd.to_datetime(business_rules_df['ImportDate'], errors='coerce')
```

```python
# Data Quality Check Example: Check for NULL values
null_counts_business_rules = business_rules_df.isnull().sum()
print("Null Counts in BusinessRules:")
print(null_counts_business_rules)
```

```python
 #Analysis 1: Count the number of unique issuers in BusinessRules
cursor.execute("SELECT COUNT(DISTINCT IssuerId) FROM BusinessRules;")
num_unique_issuers = cursor.fetchone()[0]
print(f"Number of unique issuers in BusinessRules: {num_unique_issuers}")
```

```python
# Analysis 2: Find the earliest and latest business years in BusinessRules
cursor.execute("SELECT MIN(BusinessYear) AS EarliestYear, MAX(BusinessYear) AS LatestYear FROM BusinessRules;")
business_years = cursor.fetchone()
print(f"Earliest Business Year: {business_years[0]}, Latest Business Year: {business_years[1]}")
```

```python
# Analysis 3. Identify the distribution of plan types
cursor.execute("""
    SELECT MarketCoverage, COUNT(*) AS NumPlans
    FROM BusinessRules
    GROUP BY MarketCoverage;
""")
plan_type_distribution = cursor.fetchall()
print("Distribution of plan types in BusinessRules:")
for plan_type, num_plans in plan_type_distribution:
    print(f"{plan_type}: {num_plans} plans")
```

```python
# Analysis 4 Find the average number of dependents covered in family plans

cursor.execute("""
    SELECT AVG(TwoParentFamilyMaxDependentsRule) AS AvgNumDependents
    FROM BusinessRules
    WHERE MarketCoverage = 'Family';
""")
avg_num_dependents_family_plans = cursor.fetchone()[0]
print(f"Average number of dependents covered in family plans: {avg_num_dependents_family_plans}")
```

```python
# 5. Identify distinct values in TwoParentFamilyMaxDependentsRule:
cursor.execute("""
    SELECT DISTINCT TwoParentFamilyMaxDependentsRule
    FROM BusinessRules;
""")
distinct_dependents_rules = cursor.fetchall()
print("Distinct values in TwoParentFamilyMaxDependentsRule:")
for rule_value in distinct_dependents_rules:
    print(rule_value[0])
```

```python
# 6. Identify the distribution of plan types considering text in TwoParentFamilyMaxDependentsRule:
cursor.execute("""
    SELECT MarketCoverage, COUNT(*) AS NumPlans
    FROM BusinessRules
    GROUP BY MarketCoverage, TwoParentFamilyMaxDependentsRule;
""")
plan_type_distribution = cursor.fetchall()
print("Distribution of plan types in BusinessRules:")
for plan_type, num_plans in plan_type_distribution:
    print(f"{plan_type}, Dependents Rule: {num_plans} plans")
```

```python
#7. Find the number of plans by state and market coverage
cursor.execute("""
    SELECT StateCode, MarketCoverage, COUNT(*) AS NumPlans
    FROM BusinessRules
    GROUP BY StateCode, MarketCoverage;
""")
state_market_coverage_distribution = cursor.fetchall()
print("Number of plans by state and market coverage in BusinessRules:")
for state, market_coverage, num_plans in state_market_coverage_distribution:
    print(f"State: {state}, Market Coverage: {market_coverage}, Plans: {num_plans}")
```

```python
#8. Identify distinct values in AgeDeterminationRule

cursor.execute("""
    SELECT DISTINCT AgeDeterminationRule
    FROM BusinessRules;
""")
distinct_age_determination_rules = cursor.fetchall()
print("Distinct values in AgeDeterminationRule:")
for age_rule_value in distinct_age_determination_rules:
    print(age_rule_value[0])

```

```python
#9 Count the number of plans based on AgeDeterminationRule

cursor.execute("""
    SELECT AgeDeterminationRule, COUNT(*) AS NumPlans
    FROM BusinessRules
    GROUP BY AgeDeterminationRule;
""")
plans_by_age_determination_rule = cursor.fetchall()
print("Number of plans based on AgeDeterminationRule in BusinessRules:")
for age_rule, num_plans in plans_by_age_determination_rule:
    print(f"Age Determination Rule: {age_rule}, Num Plans: {num_plans}")
```

```python
#Count the number of plans for each metal level
cursor.execute("""
    SELECT MetalLevel_2015, COUNT(*) AS NumPlans
    FROM Crosswalk2015
    GROUP BY MetalLevel_2015;
""")
plans_by_metal_level = cursor.fetchall()
print("Number of plans by metal level in Crosswalk2015:")
for metal_level, num_plans in plans_by_metal_level:
    print(f"{metal_level}: {num_plans} plans")
```

```python
#Find the average age of plans
cursor.execute("""
    SELECT AVG(AgeOffPlanID_2015) AS AveragePlanAge
    FROM Crosswalk2015;
""")
average_plan_age = cursor.fetchone()[0]
print(f"Average age of plans in Crosswalk2015: {average_plan_age:.2f} years")
```

```python
# Identify the distribution of plans by age group
cursor.execute("""
    SELECT AgeOffPlanID_2015, COUNT(*) AS NumPlans
    FROM Crosswalk2015
    GROUP BY AgeOffPlanID_2015;
""")
plans_by_age_group = cursor.fetchall()
print("Distribution of plans by age group in Crosswalk2015:")
for age_group, num_plans in plans_by_age_group:
    print(f"Age Group {age_group}: {num_plans} plans")
```

```python
# Identify the distribution of plans by state

cursor.execute("""
    SELECT State, COUNT(*) AS NumPlans
    FROM Crosswalk2015
    GROUP BY State;
""")
plans_by_state_distribution = cursor.fetchall()
print("Number of plans by state in Crosswalk2015:")
for state, num_plans in plans_by_state_distribution:
    print(f"{state}: {num_plans} plans")
```

```python
# Calculate the average and maximum age of individuals covered by plans

cursor.execute("""
    SELECT AVG(AgeOffPlanID_2015) AS AveragePlanAge, MAX(AgeOffPlanID_2015) AS MaxPlanAge
    FROM Crosswalk2015;
""")
age_summary = cursor.fetchone()
average_plan_age = age_summary[0]
max_plan_age = age_summary[1]
print(f"Average age of individuals covered by plans in Crosswalk2015: {average_plan_age:.2f} years")
print(f"Maximum age of individuals covered by plans in Crosswalk2015: {max_plan_age} years")
```

```python
#Identify the distribution of plans by metal level and age group

cursor.execute("""
    SELECT MetalLevel_2015, AgeOffPlanID_2015, COUNT(*) AS NumPlans
    FROM Crosswalk2015
    GROUP BY MetalLevel_2015, AgeOffPlanID_2015;
""")
plans_by_metal_level_and_age_group = cursor.fetchall()
print("Distribution of plans by metal level and age group in Crosswalk2015:")
for metal_level, age_group, num_plans in plans_by_metal_level_and_age_group:
    print(f"{metal_level}, Age Group {age_group}: {num_plans} plans")
```

```python
# Identify the distribution of plans by metal level and state

cursor.execute("""
    SELECT MetalLevel_2015, State, COUNT(*) AS NumPlans
    FROM Crosswalk2015
    GROUP BY MetalLevel_2015, State;
""")
plans_by_metal_level_and_state = cursor.fetchall()
print("Distribution of plans by metal level and state in Crosswalk2015:")
for metal_level, state, num_plans in plans_by_metal_level_and_state:
    print(f"{metal_level}, {state}: {num_plans} plans")

```

```python
# Identify the plans with the highest average age
cursor.execute("""
    SELECT *
    FROM Crosswalk2015
    WHERE AgeOffPlanID_2015 = (SELECT MAX(AgeOffPlanID_2015) FROM Crosswalk2015);
""")
plans_highest_average_age = cursor.fetchall()
print("Plan(s) with the highest average age in Crosswalk2015:")
for plan in plans_highest_average_age:
    print(plan)
```

```python
# Identify the plans with the maximum age

cursor.execute("""
    SELECT *
    FROM Crosswalk2015
    WHERE AgeOffPlanID_2015 = (SELECT MAX(AgeOffPlanID_2015) FROM Crosswalk2015);
""")
plans_with_maximum_age = cursor.fetchall()
print("Plan(s) with the maximum age in Crosswalk2015:")
for plan in plans_with_maximum_age:
    print(plan)
```

```python
# Identify the distribution of networks by state

cursor.execute("""
    SELECT StateCode, COUNT(DISTINCT NetworkId) AS NumNetworks
    FROM Network
    GROUP BY StateCode;
""")
networks_by_state_distribution = cursor.fetchall()
print("Distribution of networks by state in Network:")
for state, num_networks in networks_by_state_distribution:
    print(f"{state}: {num_networks} networks")
```

```python
# Find the total number of unique issuers in the Network table

cursor.execute("""
    SELECT COUNT(DISTINCT IssuerId) AS NumUniqueIssuers
    FROM Network;
""")
num_unique_issuers_network = cursor.fetchone()[0]
print(f"Number of unique issuers in Network: {num_unique_issuers_network}")
```

```python
# Identify the top 5 states with the highest number of networks

cursor.execute("""
    SELECT StateCode, COUNT(DISTINCT NetworkId) AS NumNetworks
    FROM Network
    GROUP BY StateCode
    ORDER BY NumNetworks DESC
    LIMIT 5;
""")
top_states_highest_networks = cursor.fetchall()
print("Top 5 states with the highest number of networks in Network:")
for state, num_networks in top_states_highest_networks:
    print(f"{state}: {num_networks} networks")

```

```python
service_area_df = pd.read_sql_query("SELECT * FROM ServiceArea;", conn)
```

```python
service_area_df.head()
```

```python
service_area_df.tail()
```

```python
service_area_df.info()
```

```python
# Data Transformation Example: Convert 'ImportDate' to datetime
service_area_df['ImportDate'] = pd.to_datetime(service_area_df['ImportDate'], errors='coerce')

```

```python
# Data Quality Check Example: Check for NULL values
null_counts_service_area = service_area_df.isnull().sum()

print("Null Counts in ServiceArea:")
print(null_counts_service_area)
```

```python
# Remove rows with null values
service_area_df_cleaned = service_area_df.dropna()

# Display the cleaned DataFrame
print("ServiceArea DataFrame after removing rows with null values:")
print(service_area_df_cleaned)
```

```python
# Identify the distribution of service areas by state
service_area_distribution = service_area_df_cleaned['StateCode'].value_counts()
print("Distribution of service areas by state:")
print(service_area_distribution)
```

```python
# Find the total number of unique issuers in the cleaned ServiceArea DataFrame

num_unique_issuers_service_area = service_area_df_cleaned['IssuerId'].nunique()
print(f"Number of unique issuers in ServiceArea: {num_unique_issuers_service_area}")
```

```python
#Count the occurrences of each unique 'MarketCoverage' value

market_coverage_counts = service_area_df_cleaned['MarketCoverage'].value_counts()
print("MarketCoverage value counts:")
print(market_coverage_counts)

```

```python
# Percentage distribution of 'MarketCoverage' values

market_coverage_percentage = service_area_df_cleaned['MarketCoverage'].value_counts(normalize=True) * 100
print("Percentage distribution of MarketCoverage values:")
print(market_coverage_percentage)
```

```python
# Identify unique 'MarketCoverage' values

unique_market_coverage_values = service_area_df_cleaned['MarketCoverage'].unique()
print("Unique MarketCoverage values:")
print(unique_market_coverage_values)
```

```python
# Count the number of plans for each 'MarketCoverage' category

plans_per_market_coverage = service_area_df_cleaned.groupby('MarketCoverage')['RowNumber'].count()
print("Number of plans for each MarketCoverage category:")
print(plans_per_market_coverage)
```

```python
# Distribution of plans based on 'MarketCoverage' and 'StateCode

plans_distribution_by_state = service_area_df_cleaned.groupby(['MarketCoverage', 'StateCode'])['RowNumber'].count()
print("Distribution of plans based on MarketCoverage and StateCode:")
print(plans_distribution_by_state)
```

```python
# Identify the state with the highest number of plans

state_with_highest_plans = service_area_df_cleaned.loc[service_area_df_cleaned.groupby('StateCode')['RowNumber'].idxmax()]
print("State with the highest number of plans:")
print(state_with_highest_plans[['StateCode', 'RowNumber']])
```

```python
# Count the number of unique issuers for each 'MarketCoverage' category

unique_issuers_per_market_coverage = service_area_df_cleaned.groupby('MarketCoverage')['IssuerId'].nunique()
print("Number of unique issuers for each MarketCoverage category:")
print(unique_issuers_per_market_coverage)
```
